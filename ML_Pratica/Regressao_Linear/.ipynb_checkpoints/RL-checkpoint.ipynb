## Estimativas Com Regressão Linear

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
# Criando os dados X e Y
X = np.array([50, 60, 70, 80, 100, 120]).reshape(-1, 1)  # variável independente
Y = np.array([150, 180, 200, 230, 280, 340])
```

```python
modelo = LinearRegression()
modelo.fit(X,Y)
```

```python
X_novo = np.array([[90]])
previsao = modelo.predict(X_novo)

print("Preço previsto para uma casa de 90m:", round(previsao[0],2))
```

```python
# Plotando os dados e a reta
plt.scatter(X, Y, color="blue", label="Dados reais")
plt.plot(X, modelo.predict(X), color="red", label="Regressão Linear")
plt.scatter(X_novo, previsao, color="green", label="Previsão (90m²)")
plt.xlabel("Tamanho da Casa (m²)")
plt.ylabel("Preço da Casa (mil R$)")
plt.legend()
plt.show()
```
