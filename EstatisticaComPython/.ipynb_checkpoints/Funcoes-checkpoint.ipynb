```python
def duplica_bolsa(rows):
    nova_bolsa = rows * 2
    return nova_bolsa
```

```python
df = df_prouni[['uf_busca','curso_busca','mensalidade','bolsa_integral_ampla']]
df
```

```python
df['nova_bolsa'] = df['bolsa_integral_ampla'].apply(duplica_bolsa)
```

```python
df
```

```python
#utilizando lambda para fazer a função
df['bolsa_lambda'] = df['bolsa_integral_ampla'].apply(lambda X:X*2)
```
